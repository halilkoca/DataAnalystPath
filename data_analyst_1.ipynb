```python
# 2. Get avg by genre
avg_price_by_genre = data.groupby('Genre')['Price'].mean()
print(avg_price_by_genre)
```

```python
# visualize avg by price
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x='Genre', y='Price', data=data)
plt.title('Avg Book Price by Genre')
plt.ylabel('Avg Price ($)')
plt.show()
```

```python
top_rated = data.sort_values(by='User Rating', ascending=False).head(10)
print(top_rated)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get top 10 highest-rated books
top_rated = data.sort_values(by=['User Rating', 'Reviews'], ascending=False).head(10)

# Set figure size
plt.figure(figsize=(10, 6))

# Create barplot without triggering palette warning
sns.barplot(
    x='User Rating',
    y='Name',
    data=top_rated,
    hue='Name',          # Assign hue explicitly
    palette='viridis',
    dodge=False,         # Prevent shifting due to hue
    legend=False         # Remove legend since hue == y
)

# Titles and labels
plt.title('Top 10 Highest-Rated Amazon Books (2009–2019)', fontsize=14)
plt.xlabel('User Rating')
plt.ylabel('Book Title')

plt.tight_layout()
plt.show()
```

```python
# 2. Yearly average price trends (price over time)
```

```python
# Calculate average price by year
avg_price_by_year = data.groupby('Year')['Price'].mean().reset_index()
print(avg_price_by_year)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(10, 6))

# Lineplot for yearly average prices
sns.lineplot(data=avg_price_by_year, x='Year', y='Price', marker='o', color='green')

# Add titles and labels
plt.title('Average Book Price by Year (2009–2019)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Average Price ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

3. Correlation analysis (between User Rating, Reviews, and Price)

```python
# Select revelant numeric columns
numeric_df = data[['User Rating', 'Reviews', 'Price']]

# Calculate correlation matrix
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```

```python
# Step 2: visualize correlation
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the figure
plt.figure(figsize=(6,4))

# Create heatmap
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    square=True
)

plt.title('Correlation Matrix: Rating, Reviews, Price')
plt.tight_layout()
plt.show()
```

4. Most frequent authors (best selling author count)

```python
# Count how many times each author appears
top_authors = data['Author'].value_counts().head(10)
print(top_authors)
```

```python
# Visualize the Top Authors

import matplotlib.pyplot as plt
import seaborn as sns

# Convert to DataFrame for plotting
top_authors_df = top_authors.reset_index()
top_authors_df.columns = ['Author', 'Count']

# Set figure size
plt.figure(figsize=(10,6))

# Create horizontal bar chart
sns.barplot(
    x='Count',
    y='Author',
    data=top_authors_df,
    hue='Author',         # Explicit hue
    palette='Blues_r',
    dodge=False,
    legend=False          # No need for separate legend
)

# Add titles and labels
plt.title('Top 10 Most Frequent Amazon Best Selling Authors', fontsize=14)
plt.xlabel('Number of Times listes')
plt.ylabel('Author')
plt.tight_layout()
plt.show()
```

Gary Chapman and Jeff Kinney top the list, each appearing 14 times in Amazon’s yearly Top 50 Bestsellers between 2009 and 2019. They're followed closely by Suzanne Collins (12), and Rick Riordan (11), showing the long-term commercial impact of popular book series like Diary of a Wimpy Kid and The Hunger Games. Authors like Dr. Seuss and Eric Carle show the enduring success of classic children's literature, while organizations like the APA and Gallup indicate recurring demand for non-fiction, psychological, and self-development titles.
